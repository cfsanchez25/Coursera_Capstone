# Creating a DF

```python
# setting up the lists that will form our dataframe with all the results
titles = []
prices = []
areas = []
zone = []
```

```python
%%time

headers = ({'User-Agent':
            'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36'})

n_pages = 0

for page in range(0,900):
    n_pages += 1
    bogo_rental='https://www.metrocuadrado.com/apartamentos/arriendo/'    
    r = get(bogo_rental, headers=headers)
    html_soup = BeautifulSoup(r.text, 'html.parser')
    house_containers = html_soup.find_all('div', class_="m_rs_list_item")
    if house_containers != []:
        for container in house_containers:
            
            # Price            
            price = container.find_all('span')[1].text
            price.replace('$','')
            price.replace('.','')
            price_ = [int(price[s]) for s in range(0,len(price)) if price[s].isdigit()]
            price = ''
            for x in price_:
                price = price+str(x)
            prices.append(int(price))

            # Zone
            location = container.find_all('h2')[0].text
            location=location[location.find(',')+1:]
            zone.append(location)

            # Title
            name = container. container.find_all('h2')[0].text
            titles.append(name)

            # Area
            m2 = container.find_all('span')[3].text
            m2.replace('m2','')
            m2 = float(''.join(itertools.takewhile(str.isdigit, m2)))
            areas.append(m2)
    else:
        break
    
    sleep(randint(1,2))
    
print('You scraped {} pages containing {} properties.'.format(n_pages, len(titles)))
```
